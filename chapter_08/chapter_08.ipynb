## Square root transform

```python
series = [i**2 for i in range(1,100)]

plt.figure(1)

plt.subplot(211)
plt.plot(series)

plt.subplot(212)
plt.hist(series)

plt.show()
```

```python
transform = np.sqrt(series)

plt.figure(1)

plt.subplot(211)
plt.plot(transform)

plt.subplot(212)
plt.hist(transform)

plt.show()
```

```python
series = pd.read_csv('airline-passengers.csv', header=0, index_col=0, parse_dates=True)

series.columns = ['passengers']
series['passengers'] = np.sqrt(series['passengers'])

plt.figure(1)

plt.subplot(211)
plt.plot(series['passengers'])

plt.subplot(212)
plt.hist(series['passengers'])

plt.show()
```
