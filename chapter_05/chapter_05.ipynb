```python
import pandas as pd
```

```python
series = pd.read_csv('daily-min-temperatures.csv', header=0, index_col=0, parse_dates=True)
series
```

```python
df = pd.DataFrame()
df['month'] = [series.index[i].month for i in range(len(series))]
df['day'] = [series.index[i].day for i in range(len(series))]
df['temperature'] = series['Temp'].values
df.head()
```

```python
temps = pd.DataFrame(series['Temp'].values)
temps.head()
```

## Lag Features

```python
dflag = pd.concat([temps.shift(1), temps], axis=1)
dflag.columns = ['t', 't+1']
dflag.head()
```

## Lag Features size 3 - shift

```python
dflag3 = pd.concat([temps.shift(3), temps.shift(2), temps.shift(1), temps], axis=1)
dflag3.columns = ['t-2', 't-1', 't', 't+1']
dflag3.head()
```

## Rolling Window

```python
shifted = temps.shift(1)
window = shifted.rolling(window=2)
means = window.mean()
means.head()
```

```python
dfrol = pd.concat([means, temps], axis=1)
dfrol.columns = ['mean(t-1,t)', 't+1']
dfrol.head()
```

## Rolling Window 3 size

```python
width = 3
shifted = temps.shift(width - 1)
window = shifted.rolling(window = width)
dfrol3 = pd.concat([window.min(), window.mean(), window.max(), temps], axis=1)
dfrol3.columns = ['min', 'mean', 'max', 't+1']
dfrol3.head()
```

## Expantion

```python
window = temps.expanding()
dfexp = pd.concat([window.min(), window.mean(), window.max(), temps.shift(-1)], axis=1)
dfexp.columns = ['min', 'mean', 'max', 't+1']
dfexp.head()
```
