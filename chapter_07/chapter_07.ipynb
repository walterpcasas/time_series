## Upsampling

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')
```

```python
series = pd.read_csv('shampoo.csv', header=0, index_col=0, parse_dates=True, date_parser=parser)
series.head()
```

```python
upsampled = series.resample('D').mean()
upsampled.head(20)
```

```python
upsampled = series.resample('D').mean()
interpolated = upsampled.interpolate(method='linear')
interpolated.head(20)
```

```python
interpolated.plot()
plt.show()
```

```python
upsampled = series.resample('D').mean()
interpolated = upsampled.interpolate(method='spline', order=2)
interpolated.head(20)
```

```python
interpolated.plot()
plt.show()
```

## Downsampling

```python
resample = series.resample('Q').mean()
resample.head(20)
```

```python
resample.plot()
plt.show()
```

```python
resample = series.resample('A').sum()
resample.head(20)
```

```python
resample.plot()
plt.show()
```
